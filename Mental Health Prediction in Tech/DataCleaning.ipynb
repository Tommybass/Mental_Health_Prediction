```python
# create a function to fill null values with None
def fillnull(df):
    df['How many employees does your company or organization have']=df['How many employees does your company or organization have'].fillna('None')
    df['If yes, what conditions have you been diagnosed with']=df['If yes, what conditions have you been diagnosed with'].fillna('None')
    df['If maybe, what conditions do you believe you have']=df['If maybe, what conditions do you believe you have'].fillna('None')
    df['If so, what conditions were you diagnosed with'] = df['If so, what conditions were you diagnosed with'].fillna('None')
fillnull(df)
```

```python
def clean_df(df):
    for index , row in enumerate(df['If so, what conditions were you diagnosed with']):
        if row == 'None':
            df['If so, what conditions were you diagnosed with'][index] = df['If yes, what conditions have you been diagnosed with'][index]
            df['If so, what conditions were you diagnosed with'][index] = df['If maybe, what conditions do you believe you have'][index]
clean_df(df)
```

```python
# Set a visually appealing Seaborn style
sns.set_style("whitegrid")

# Create a figure with a more descriptive size
plt.figure(figsize=(10, 6))

# Calculate the percentage distribution of 'Age Group'
eda_percentage = df['Age Group'].value_counts(normalize=True).rename_axis('Age Group').reset_index(name='Percentage')

# Sort the data by Percentage in descending order for better visualization
eda_percentage = eda_percentage.sort_values(by='Percentage', ascending=False).head(10)

# Create a horizontal bar plot using Seaborn
sns.barplot(x='Percentage', y='Age Group', data=eda_percentage, palette='viridis')

# Set the title to be more informative
plt.title('Top 10 Age Groups Distribution in the Dataset', fontsize=16)

# Set axis labels
plt.xlabel('Percentage', fontsize=12)
plt.ylabel('Age Group', fontsize=12)

# Show the plot
plt.show()
```
